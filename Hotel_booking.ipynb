```python
def filter_numbers(x):
    return [i for i in x if type(i) in (int,float)]

filter_numbers([1,2,'A',[1,2],True])
```

Q2:
Viết hàm count_element_types(x) nhập vào 1 list x, trả về 1 tuple có 4 phân tử:
1. Số các phần tử là số dương trong x
2. Số các phần tử là số âm trong x
3. Số các phần tử là số 0 trong x
4. Số các phần tử không phải là số trong x

```python
def count_element_types(x):
    a=b=c=d=0
    for i in x:
        if type(i) in (int,float) == False:
            if int(i) > 0:
                a += 1
            elif int(i) < 0:
                b += 1
            else:
                c += 1
        else:
            d += 1
    return (a,b,c,d)
count_element_types([1,2,'A',[1,2],True,-1,-2.5])
```

Viết hàm get_quaters(x) nhập vào một số nguyên từ 1 đến 12 thể hiện số thứ tự của tháng trong năm trả về số nguyên là quater tương ứng từ 1 đến 4
Ví dụ: tháng 1,2,3 tương ứng số quater 1, tháng 4,5,6 tương ứng với số quater 2, ...

```python
def get_quaters(x):
    if x in [1,2,3]:
        return 1
    elif x in [4,5,6]:
        return 2
    elif x in [7,8,9]:
        return 3
    elif x in [10,11,12]:
        return 4
    else:
        return 0
for x in range(1,13): print(get_quaters(x), end=" ")
        
```

```python
Q3: Thay vì tạo list các cột có giá trị null một cách manual, xây dựng hàm null_count(), nhận vào tên tập data và trả ra dataframe chứa tên các cột bị null và số lượng giá trị null và % null
```

```python
def null_count(data_file):
  """
  Hàm tính số lượng và tỷ lệ giá trị null trong từng cột của DataFrame.

  Args:
    data_file: Tên file chứa dữ liệu.

  Returns:
    DataFrame: Chứa tên cột, số lượng giá trị null và tỷ lệ phần trăm giá trị null.
  """

  # Đọc dữ liệu từ file
  df = pd.read_csv(data_file)

  # Tính số lượng giá trị null cho mỗi cột
  null_counts = df.isnull().sum()

  # Tính tổng số hàng
  total_rows = len(df)

  # Tính tỷ lệ phần trăm giá trị null
  null_percentage = (null_counts / total_rows) * 100

  # Tạo DataFrame kết quả
  null_summary = pd.DataFrame({'Số lượng null': null_counts,
                               'Tỷ lệ % null': null_percentage})
  #null_summary[null_summary['Số lượng null'] > 0] == True
  
  #return null_data
  return null_summary[null_summary['Số lượng null'] > 0]

# Ví dụ sử dụng hàm
result = null_count('hotel_bookings.csv')
print(result)
```

```python
def null_count_v2(x):
    count_null = x.isnull().sum().sort_values(ascending = True)
    percent_null = (x.isnull().sum()/len(x)*100)
    null_data = pd.concat([count_null,percent_null], axis = 1, keys = ['Số lượng null','Tỷ lệ % null'])
    null_data = null_data[null_data['Số lượng null']>0]
    return null_data
null_count_v2(hotel)
```

```python
Q2: Vẽ pie chart thể hiện % của các loại meal hoặc meal_service
```

```python
def pie(data, label, col, title = None, figsize = (12,8), fontsize = 14):
    plt.figure(figsize = figsize)
    labels = data[label]
    plt.pie(x = data[col], autopct = "%.1f%%", explode = [0.05]*data[label].count(), labels = labels, pctdistance = 0.5)
    plt.title(title, fontsize = fontsize)
```

```python
df = hotel.groupby('meal_service').size().reset_index(name='count')
pie(data = df, label = 'meal_service', col = 'count')
plt.title('Meal service contribution')
```
